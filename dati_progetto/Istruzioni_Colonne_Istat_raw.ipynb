# Dati per il progetto di Introduzione alla Programmazione

In questo archivio sono contenuti due dataset: un dataset di ISTAT che contiene informazioni sugli occupati in vari settori lavorativi e PAISÀ, un corpus di testi reperiti online in italiano.

## PAISÀ

Nel file `data/paisa/paisa.csv` è contenuta una parte dell'intero corpus [PAISÀ](www.corpusitaliano.it). Il formato del file è CSV e contiene tre colonne:

- id: un id assegnato a ciascun documento nel corpus
- text: il contenuto testuale di ciascun documento
- url: l'indirizzo (URL) da cui è stato reperito il documento

## ISTAT

Il dataset ISTAT nel file `data/istat/dataset.csv` contiene informazioni su occupati e settori lavorativi. Una parte dei dati di questo dataset può essere visualizzata [qui](https://esploradati.istat.it/databrowser/#/it/dw/categories/IT1,Z0900ENT,1.0/ENT_STRU/DICA_ACDP/DICA_ACDP_REG/IT1,183_203_DF_DICA_ACDP_9,1.0).
Ogni riga del file è un "data point", in cui sono stati variati uno o più dei parametri.
Contiene le seguenti righe:

- FREQ: la frequenza con la quale viene memorizzato questo dato. Nel nostro caso è sempre "A" che sta per "annuale"
- REF_AREA: l'area geografica a cui si riferisce il dato in formato di codice. Nel file CSV `data/istat/columns/REF_AREA.csv` vi ho fornito tutti i codici per poterla decodificare.
- DATA_TYPE: il tipo di dato memorizzato. Nel nostro caso questa colonna contiene codici che rappresentano diversi tipi di impiego (lavoratori dipendenti, indipendenti, esterni, temporanei). Nel file `data/istat/columns/DATA_TYPE.csv` ho inserito le descrizioni dei codici per poterli interpretare.
- ECON_ACTIVITY_NACE_2007: in questo campo sono contenuti i codici ATECO che indicano diverse tipologie di lavoro. I codici ATECO (escludendo 010 che rappresenta il totale) vanno da valori generici a valori via via più specifici con l'aumento del numero di cifre. Nel file `data/istat/columns/ECON_ACTIVITY_NACE_2007.csv` ci sono tutti le corrispondenze fra i codici ATECO e la loro descrizione.
- PERS_EMPL_SIZE_CLASS: una colonna che contiene informazioni sulla dimensione delle aziende. Nel dataset è sempre TOTAL.
- LEGAL_FORM: classificazione legale degli impiegati (dirigenti, professioni tecniche, etc). Nel dataset è sempre TOT.
- AGE_ENTERPRISE: l'età dell'impresa. Nel dataset è sempre TOTAL.
- SEX: il sesso dei lavoratori. Nel file `data/istat/columns/SEX.csv` sono documentati i valori possibili.
- AGE: l'età dei lavoratori. Nel dataset è sempre Y_GE15, che indica età maggiori e uguali a 15 anni.
- COUNTRY_BIRTH: il paese di origine dei lavoratori. Nel dataset è sempre WORLD.
- 2012, 2013, 2014, 2015, 2016, 2017: contengono il numero di occupati medi in un certo anno. **ATTENZIONE: non tutti i dati sono presenti per tutti gli anni!**
